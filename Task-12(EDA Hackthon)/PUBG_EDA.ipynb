#**Detect Outliers**

* **outliers are observations that lie far away from the majority of the observations in the data set and can be represented mathematically in different ways**
* **one method of defining outliers are:The points lying beyond the (third quartile+1.5 inter quartile range) and below (first quartile -1.5 IQR)**


```python
df.plot(kind='box',figsize=(16,16),layout=(8,6),sharex=False,subplots=True)
```

```python
def catch_data(df):
  for col in df.columns:
      print("catch column",col)
      if (((df[col].dtype)=='float64')|((df[col].dtype)=='int64')):
            percentiles=df[col].quantile([0.01,0.99]).values
            df[col][df[col] <= percentiles[0]]=percentiles[0]
            df[col][df[col] >= percentiles[1]]=percentiles[1]
      else:
        df[col]=df[col]
  return df
final_df=catch_data(df)
```

```python
df.plot(kind='box',figsize=(16,16),layout=(8,6),sharex=False,subplots=True)
```

```python
plt.boxplot(df['swimDistance'])
plt.title('Boxplot before removing the outliers')
plt.show()
def drop_outliers(df,field_name):
  iqr=1.5*(np.percentile(df[field_name],75)-np.percentile(df[field_name],25))
  df.drop(df[df[field_name]>(iqr+np.percentile(df[field_name],75))].index,inplace=True)
  df.drop(df[df[field_name]<(np.percentile(df[field_name],25)-iqr)].index,inplace=True)
drop_outliers(df,'swimDistance') 
plt.boxplot(df['swimDistance']) 
plt.title('Box plot after removing outliers')
plt.show()
```

**checking the pearson correaltion between variables**

```python
k=5
#here i want to take 5 number of variables for heatmap
f,ax=plt.subplots(figsize=(11,11))
cols=df.corr().nlargest(k,'winPlacePerc')['winPlacePerc'].index
cm=np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm=sns.heatmap(cm,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size':10},yticklabels=cols.values,xticklabels=cols.values)
plt.show()
```
